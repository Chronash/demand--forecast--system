```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

PLOTS_DIR = Path("../artifacts/plots")
PLOTS_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_csv("../artifacts/generated/sales_daily.csv")
df["sale_date"] = pd.to_datetime(df["sale_date"])

print(f"Строк: {len(df):,}")
print(f"Столбцов: {len(df.columns)}")
print(df.dtypes)
```

```python
df.describe()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df["demand_units"], bins=60, kde=True, color="#2196F3")
plt.title("Распределение спроса (demand_units)", fontsize=14)
plt.xlabel("Спрос (штук)")
plt.ylabel("Частота")
plt.tight_layout()
plt.savefig(PLOTS_DIR / "demand_distribution.png", dpi=150)
plt.show()
```

```python
monthly = df.groupby("month")["demand_units"].mean().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly, x="month", y="demand_units", marker="o", linewidth=2.5)
plt.title("Средний спрос по месяцам", fontsize=14)
plt.xlabel("Месяц")
plt.ylabel("Средний спрос (штук)")
plt.xticks(range(1, 13))
plt.tight_layout()
plt.savefig(PLOTS_DIR / "demand_by_month.png", dpi=150)
plt.show()
```

```python
sample = df.sample(min(15000, len(df)), random_state=42)

plt.figure(figsize=(12, 5))
sns.scatterplot(data=sample, x="temperature", y="demand_units", alpha=0.3, s=15)
plt.title("Влияние температуры на спрос", fontsize=14)
plt.xlabel("Температура (°C)")
plt.ylabel("Спрос (штук)")
plt.tight_layout()
plt.savefig(PLOTS_DIR / "temp_impact.png", dpi=150)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
region_order = df.groupby("region")["demand_units"].mean().sort_values(ascending=False).index
sns.boxplot(data=df, x="region", y="demand_units", order=region_order)
plt.title("Распределение спроса по регионам", fontsize=14)
plt.xlabel("Регион")
plt.ylabel("Спрос (штук)")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ["month", "weekday", "day_of_year", "is_weekend",
                "is_holiday", "promo_flag", "temperature", "price", "demand_units"]
corr = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Матрица корреляций", fontsize=14)
plt.tight_layout()
plt.show()
```
